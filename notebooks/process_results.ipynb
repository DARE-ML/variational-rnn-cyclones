```python
import os
import pandas as pd
```

```python
runs = os.listdir('../runs/')

results = []

for i, run in enumerate(runs):
    result_json = os.path.join('../runs', run, 'results.json')
    try:
        results.append(pd.read_json(result_json, orient='records'))
    except:
        pass

results_df = pd.concat(results).sort_values(by=['ds_name', 'model']).reset_index(drop=True)
```

```python
results_df.loc[:, 'train_mse_mean']= results_df.loc[:, 'train_mse'].apply(lambda x: x['mean'])
results_df.loc[:, 'train_mse_std']= results_df.loc[:, 'train_mse'].apply(lambda x: x['std'])

results_df.loc[:, 'test_mse_mean']= results_df.loc[:, 'test_mse'].apply(lambda x: x['mean'])
results_df.loc[:, 'test_mse_std']= results_df.loc[:, 'test_mse'].apply(lambda x: x['std'])
```

```python
results_summary = results_df.loc[:, ['run_name', 'ds_name', 'model', 'train_mse_mean', 'train_mse_std', 'test_mse_mean', 'test_mse_std', 'energy_score']]

results_summary_intensity = results_summary.loc[results_summary.run_name.str.contains('intensity')].drop(columns=['run_name'])

results_summary_intensity
```

```python
results_summary_location = results_summary.loc[results_summary.run_name.str.contains('location')].drop(columns=['run_name'])

results_summary_location
```
